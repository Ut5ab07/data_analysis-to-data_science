# Merge, Join, Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/LOTR.csv")
df1
```

```python
df2 = pd.read_csv("data/LOTR 2.csv")
df2
```

##### MERGE


```python
df1.merge(df2, how ='in ner', on = ['FellowshipID','FirstName']) #inner-join
```

```python
df1.merge(df2, how = 'inner') #not merged on any specific columns
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

##### Join


```python
df1.join(df2, on= 'FellowshipID', how='outer',lsuffix='_Left',rsuffix='_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),
                                         lsuffix='_Left',rsuffix='_Right',how ='outer')
df4
```

##### Concatenate

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'outer', axis =1)
```

```python
pd.concat([df1,df2], join = 'inner')
```
