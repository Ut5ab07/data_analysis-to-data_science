##### Cleaning column names

```python
df.columns =(df.columns
            .str.strip()
            .str.lower()
             .str.replace(' ','_')
    )
df
```

```python
df.isna().sum()
```

##### arranging values based on continent to fill the missing continent

```python
df= df.sort_values(by='continent')
df
```

##### Filling the missing datas- median for numerical values, front fill and back fill for missing continents and replacing NaN by unknown for other missing categorial values 

```python
num_cols= df.select_dtypes(include='number').columns
df[num_cols]=df[num_cols].fillna(df[num_cols].median())
```

```python
df['continent']=df['continent'].ffill().bfill()
```

```python
cat_cols= df.select_dtypes(include='object').columns
df[cat_cols]=df[cat_cols].fillna('unknown')
```

```python
df.isna().sum() #No more missing values
```

##### Checking and removing duplicates

```python
df.duplicated().sum() #No duplicate values, so no cleaning required
```

##### Cleaning Texts

```python
df['country']=(df['country']
               .str.strip()
               .str.lower()
               .str.replace(r'^a-zA-Z,-','',regex=True) 
               .str.replace(r'[!@#$%^&*;]','',regex=True)
)
```

```python
df
```

##### Fixing datatypes

```python
df['day']=pd.to_datetime(df['day'], errors='coerce')
```
