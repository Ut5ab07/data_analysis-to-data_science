# Data Cleaning in Pandas


```python
import pandas as pd
```

```python
df = pd.read_excel("data/Customer Call List.xlsx")
df
```

##### removing duplicates

```python
df = df.drop_duplicates()
df
```

##### Dropping Columns

```python
df=df.drop(columns="Not_Useful_Column")
df
```

##### Strip
##### Can only be used if something needs to be removed from both ends

```python
df["Last_Name"] =df["Last_Name"].str.lstrip("...")
df["Last_Name"] =df["Last_Name"].str.lstrip("/")
df["Last_Name"] =df["Last_Name"].str.rstrip("_")
df

#lstrip-> for left side
#rstrip-> for right side
```

```python
df = pd.read_excel("data/Customer Call List.xlsx")
df
df=df.drop(columns="Not_Useful_Column")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.loc[:,"Last_Name"]=df["Last_Name"].str.strip("123._/")
df
```

##### Cleaning and standardizing Phone number

```python
df.loc[:,"Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df
#Replaces everything except characters fro a-z, A-Z and 0-9
```

```python
df.loc[:,"Phone_Number"]=df["Phone_Number"].apply(lambda x : str(x)) 
#Converts entire data of the column into string
df.loc[:,"Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df.loc[:,"Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df.loc[:,"Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
df
```

##### Splitting Columns

```python
df.loc[:,['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',n=1,expand=True)
df
```

##### Standardizing Column Data

```python
df.loc[:,'Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df.loc[:,'Paying Customer']=df['Paying Customer'].str.replace('No','N')
df
```

```python
df.loc[:,'Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df.loc[:,'Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
df
```

```python
df=df.fillna('') #Nan-> Not a number will be replaced by white space
df
```

##### Filtering Down the rows of data

```python
for x in df.index:
    if df.loc[x,"Phone_Number"]=='':
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]== 'Y':
        df.drop(x,inplace=True)

df
```

```python
df=df.reset_index(drop=True)
df
```
