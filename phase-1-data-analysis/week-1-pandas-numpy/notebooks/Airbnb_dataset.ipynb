# Airbnb Real World Data Cleaning

```python
import pandas as pd
```

##### Loading dataset

```python
df= pd.read_csv("data/Airbnb_Open_Data.csv",low_memory=False)
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns=(
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ','_')
)
df
```

##### Checking columns with NaN values

```python
df.isna().sum().sort_values(ascending=False)
```

##### Dropiing column with most NaN and less significance

```python
df= df.drop(columns=['license'])
df.info()
```

##### Handling NaN values

NaN in Numbers handled by filling median value and in Texts(objects) by replacing it with "Unknown"

```python
nums_cols = df.select_dtypes(include='number').columns
df[nums_cols]=df[nums_cols].fillna(df[nums_cols].median())
```

```python
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols]=df[cat_cols].fillna('Unknown')
```

```python
df.info() #No more null values
```

##### Cleaning Text Columns

```python
text_cols=['name','host_name','neighbourhood_group','neighbourhood']

for col in text_cols:
    df[col]=(
        df[col]
        .astype(str)
        .str.strip()
        .str.lower()
        .str.replace(r'[^a-z]',' ',regex=True)
        .str.replace(r'\s+',' ',regex=True)
    )
```

```python
df.head()
```

##### Cleaning Price and Service Fee Columns

```python
df['price'] = (
    df['price']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)
    .pipe(pd.to_numeric, errors='coerce')
)

df['service_fee'] = (
    df['service_fee']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)
    .pipe(pd.to_numeric, errors='coerce') #converts string to float
)
```

```python
df.head()
```

##### Checking duplicated values

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates() #removing duplicates
```

```python
df.duplicated().sum()
```

##### Handling outliers

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR= Q3-Q1

df=df[(df['price'] >=Q1-1.5*IQR) &
    (df['price'] <= Q3+1.5*IQR)]
```

```python
df.info()
```

```python
df.isna().sum() #Still NaN values in service_fee
```

```python
df['service_fee']=df['service_fee'].fillna(df['service_fee'].median())
df['service_fee']
```

```python
df.isna().sum() #No more NaN values
```

```python
df.describe()
```

##### Finally, data ready for further processing  

##### Saving clean data

```python
df.to_csv('Cleaned_Airbnb_data.csv', index=False)
```
