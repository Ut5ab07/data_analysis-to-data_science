##### 25. Drop columns with more than 30% missing values. 

```python
miss_percent=df.isna().sum()/len(df)*100
cols=miss_percent[miss_percent>30.0].index
df_filtered= df.drop(columns= cols)
df_filtered
```

##### 26. Fill missing numerical values with the column mean.

```python
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].mean())
df['Total Spent'].isna().sum()
```
