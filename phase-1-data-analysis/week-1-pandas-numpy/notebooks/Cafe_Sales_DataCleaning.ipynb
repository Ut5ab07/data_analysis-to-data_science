```python
df.columns=(
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ','_')
)
```

```python
df
```

```python
df.isna().sum()
```

```python
cat_cols= df.select_dtypes(include='object').columns
df[cat_cols]=df[cat_cols].fillna('UNKNOWN')
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['transaction_date']= pd.to_datetime(df['transaction_date'], errors='coerce')
```

```python
df['price_per_unit']=df['price_per_unit'].replace('UNKNOWN','')
df['total_spent']=df['total_spent'].replace('ERROR','')
```

```python
df['price_per_unit']=pd.to_numeric(df['price_per_unit'], errors='coerce')
```

```python
df['total_spent']=pd.to_numeric(df['total_spent'], errors='coerce')
```

```python
df['quantity']=pd.to_numeric(df['quantity'],errors='coerce')
```

```python
df.dtypes
```

```python
df
```

```python
cols=['transaction_id', 'item', 'quantity', 'price_per_unit', 'total_spent',
       'payment_method', 'location', 'transaction_date']
df[cols].nunique()
```

```python
df['location'].dropna().unique()
```

```python
df['payment_method'].dropna().unique()
```

```python
df['location']=df['location'].replace(['UNKNOWN','ERROR'],'N/A')
df['payment_method']=df['payment_method'].replace(['UNKNOWN','ERROR'],'N/A')
```

```python
df[cols].nunique()  #item has issues as there are only 8 items in menu
```

```python
df['item'].dropna().unique()
```

```python
df['item']=df['item'].replace(['UNKNOWN','ERROR'],'N/A')
```

```python
df[cols].nunique() 
```

```python
df.isna().sum()
```

```python
#Replacing NaN values by median of available values
num_cols= df.select_dtypes(include='number').columns
df[num_cols]=df[num_cols].fillna(df[num_cols].median())
```

```python
df.isna().sum()
```

```python
date_cols= df.select_dtypes(include='datetime').columns
df[date_cols]=df[date_cols].ffill().bfill()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
Q1 = df['total_spent'].quantile(0.25)
Q3 = df['total_spent'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['total_spent'] < lower) | (df['total_spent'] > upper)]
outliers.shape
```

```python
df.to_csv('CLeaned_Cafe_SalesData.csv',index=False)
```
