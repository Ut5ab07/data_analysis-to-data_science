# Pandas Basics

##### Pandas Series

```python
#Creating the series
import pandas as pd

a=[1,7,2]
var = pd.Series(a)
print(var)
```

```python
#Labels
print(var[0]) #returns first value of the series
```

```python
#Creating labels
import pandas as pd

a=[1,5,8]
var = pd.Series(a, index=["x","y","z"])
print(var)
```

```python
#Key-Value Objects as Series
import pandas as pd

calories ={"day1":420, "day2":460, "day3":390}
var = pd.Series(calories)
print(var)
```

```python
#Creating series using only the selected data
import pandas as pd
calories = {"day1":420, "day2":460, "day3":390}
var = pd.Series(calories, index= ["day1","day2"])
print(var)
```

##### Pandas DataFrames


```python
#Creating DataFrames

data={
    "calories": [420,460,390],
    "duration": [50,40,45]
}
df = pd.DataFrame(data)
print(var)
```

```python
#Locating Row
print(df.loc[0]) #Returns Pandas Series
```

```python
print(df.loc[[0,1]]) #Using [], Returns Pandas DataFrame 
```

```python
#Named Indexes
import pandas as pd

data = {
    "calories" : [420,460,390],
    "duration" : [50,40,45]
}
df = pd.DataFrame(data, index=["Day1","Day2","Day3"])
print(df)
```

```python
#Locating Named Indexes
print(df.loc["Day1"]) #Returns Pandas Series
```
