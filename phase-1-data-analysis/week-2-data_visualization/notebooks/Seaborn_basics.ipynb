##### HeatMaps

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
import pandas as pd

titanic['sex'] = titanic['sex'].apply(lambda x: 1 if x == 'male' else 0)
titanic['alive'] = titanic['alive'].apply(lambda x: 1 if x == 'yes' else 0)
titanic['alone'] = titanic['alone'].astype(int)

class_type = pd.CategoricalDtype(
    categories=['First', 'Second', 'Third'],
    ordered=True
)

titanic['class'] = titanic['class'].astype(class_type).cat.codes

titanic = titanic.drop(
    ['embarked', 'who', 'adult_male', 'deck', 'embark_town'],
    axis=1
)
```

```python
titanic
```

```python
titanic.corr()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
sns.heatmap(titanic.corr(), annot=True, cmap='coolwarm')
```
