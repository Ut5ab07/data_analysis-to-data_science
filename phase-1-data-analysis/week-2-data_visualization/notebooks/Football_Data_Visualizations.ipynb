```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Dataset

```python
df = pd.read_csv(r'data/player_stats.csv', encoding='latin1')
df
```

```python
df.head()
```

```python
df.tail()
```

#### Preparing Data for Visualization

```python
df.info()
```

```python
#Fixing column names to maintain uniformity
df.columns = (df.columns
    .str.strip()
    .str.lower()
    .str.replace(r'[^a-z0-9_]+','_', regex=True)
)
df
```

```python
df = df.rename(columns={
    'team' : 'country'
})
df
```

```python
df.isna().sum()
```

```python
df['club']=df['club'].fillna('No Club')
```

```python
num_cols = df.select_dtypes(include= 'number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
df.isna().sum()
```

```python
#selecting only those player that played more than 350 minutes i.e. at least qualified for knockouts
eligible_player= df[df['minutes']>=350]
eligible_player
```

```python
defender = eligible_player[eligible_player['position'] == 'DF' ]
defender
```

```python
midfielder = eligible_player[eligible_player['position'] == 'MF']
midfielder
```

```python
forward = eligible_player[eligible_player['position'] == 'FW']
forward
```

```python
goalie = eligible_player[eligible_player['position'] == 'GK']
goalie
```

## Visualizing data

##### defenders with most minutes

```python
mins_defender = defender.sort_values('minutes', ascending=False).head(5)
mins_defender
```

```python
plt.figure(figsize=(10,5))
plt.bar(mins_defender['player'], mins_defender['minutes'], width=0.4)
plt.xticks(rotation=45)
plt.title('Defenders with most minutes played')
plt.xlabel('Players')
plt.ylabel('Minutes Played')
plt.show()
```

##### Goalkeepers with most minutes per 90

```python
mins_goalie = goalie.sort_values('minutes_90s', ascending=False).head(5)
mins_goalie
```

```python
plt.figure(figsize=(10,5))
plt.bar(mins_goalie['player'],mins_goalie['minutes_90s'], width=0.4, color= 'purple')
plt.xticks(rotation=45)
plt.title('Top 5 Goalkeepers with most minutes per 90 played')
plt.xlabel('Players')
plt.ylabel('Minutes per 90')
plt.show()
```

```python
df.columns
```

##### Midfielders with most Goals, Assists and Penalty per 90

```python
elg_mid = midfielder.sort_values('goals_assists_pens_per90', ascending=False).head(5)
elg_mid
```

```python
plt.figure(figsize=(10,5))
plt.bar(elg_mid['player'], elg_mid['goals_assists_pens_per90'], width=0.4, color ='green')
plt.xticks(rotation=45)
plt.title('Top 5 midfielders with most Goal, Assists and Penalties per 90')
plt.xlabel('Players')
plt.ylabel('Goals, Assists, Penalties per 90')
plt.show()
```

##### Forwards with most goals and penalties per 90

```python
elg_fwd = forward.sort_values('goals_pens_per90', ascending=False).head(5)
elg_fwd
```

```python
plt.figure(figsize=(10,5))
plt.bar(elg_fwd['player'],elg_fwd['goals_pens_per90'], width=0.4, color='orange')
plt.xticks(rotation=45)
plt.title('Top 5 forwards with most Goals and Penalties per 90')
plt.xlabel('Players')
plt.ylabel('Goals and penalties per 90')
plt.show()
```

### Creating Heatmap

```python
import seaborn as sns
```

```python
df.columns
```

```python
df_col = ['games',
       'games_starts', 'minutes', 'minutes_90s', 'goals', 'assists',
       'goals_pens', 'pens_made', 'pens_att', 'cards_yellow', 'cards_red',
       'goals_per90', 'assists_per90', 'goals_assists_per90',
       'goals_pens_per90', 'goals_assists_pens_per90', 'xg', 'npxg',
       'xg_assist', 'npxg_xg_assist', 'xg_per90', 'xg_assist_per90',
       'xg_xg_assist_per90', 'npxg_per90', 'npxg_xg_assist_per90']
df_subset= df[df_col]
```

```python
df_corr = df_subset.corr(numeric_only=True)
df_corr
```

```python
plt.figure(figsize=(20,15))
plt.title('Heamap showing correlation between all the numeric values')
sns.heatmap(df_corr, annot=True)
plt.show()
```
