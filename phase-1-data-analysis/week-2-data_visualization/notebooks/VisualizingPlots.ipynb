#### Histograms

```python
ages = np.random.normal(20,1.5,1000)
print(ages)
```

```python
plt.hist(ages);
```

```python
plt.hist(ages,
        bins=20);
```

```python
plt.hist(ages,
        bins=[ages.min(),18,21,ages.max()]);
```

```python
#Cumulative Histogrmas
plt.hist(ages,
        bins=20,
        cumulative =True);
#shows max age is near 24
```

#### Pie Charts

```python
lans = ['C++','C#','Python','Java','JavaScript']
votes= [50,24,14,6,17]
```

```python
plt.pie(votes, labels=lans)
plt.show()
```

```python
explodes=[0,0,0,0.3,0]
plt.pie(votes, labels=lans, explode = explodes,
       autopct='%.2f%%', pctdistance=1.4,
       startangle = 90)
plt.show()
```

#### Customizing Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
years=[2014,2015,2016,2017,2018,2019,2020,2021]
income=[55,56,62,62,72,72,73,75]
income_ticks = list(range(50,81,2))
```

```python
plt.plot(years,income)
plt.title('Income of John', fontsize=24, fontname='Times New Roman')
plt.xlabel('Year')
plt.ylabel('Yearly Income')
plt.yticks(income_ticks,[f'{x}k' for x in income_ticks])
plt.show()
```

#### Adding Legends

#### Multiple figures

#### Subplots

```python
x = np.arange(100)
fig,axs = plt.subplots(2,2)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title('Sine Wave')

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title('Cosine Wave')

axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title('Random Function')

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title('Log Function')

fig.suptitle('Multiple Plots')
plt.tight_layout()
plt.show()
```

#### 3D Plotting

```python
ax = plt.axes(projection='3d')

x= np.random.random(100)
y= np.random.random(100)
z= np.random.random(100)

ax.scatter(x,y,z)
ax.set_title('3D Plot')
ax.set_xlabel('test')
plt.show()
```

```python
ax = plt.axes(projection='3d')

x= np.arange(-5,5,0.1)
y= np.arange(-1,5,0.1)

X, Y = np.meshgrid(x,y)

Z= np.sin(X)* np.cos(Y)

ax.plot_surface(X,Y,Z, cmap='Spectral')
ax.set_title('3D Plot')
ax.set_xlabel('test')
#ax.view_init(azim=0, elev=90)
plt.show()
```
